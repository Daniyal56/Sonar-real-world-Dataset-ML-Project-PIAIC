```python
import tensorflow as tf
import keras
from keras import layers
from keras.layers import Dense
import pandas
import numpy
from numpy import array
```

```python
dataframe = pandas.read_csv("sonar.csv", header=None) 
dataset = dataframe.values 
# split into input (X) and output (Y) variables 
X = dataset[:,0:60].astype(float) 
Y = dataset[:,60] 
```

```python
dataframe
```

```python
dataframe[60] = [1 if x == 'R' else 0  for x in dataframe[60]] 
```

```python
y = dataframe[60]
y.values
```

```python
class ML(keras.Model):
    
    def __init__(self):
        super(ML,self).__init__()
        inputs = keras.Input=(60,)
        self.L1 = Dense(30, activation='relu')
        self.L2 = Dense(30, activation='relu')
        self.L3 = Dense(10, activation='sigmoid')
        
    def call(self,inputs):
        x = self.L1 (inputs)
        x= self.L2 (x)
        return self.L3(x)
```

```python
model = ML()
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
model.fit(X,y, epochs=100,batch_size=32)
```
